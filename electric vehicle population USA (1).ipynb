## Relationship Analysis

```python
corelation = Electric_Vehicle_Population_Data.corr()
```

```python
sns.heatmap(corelation,xticklabels=corelation.columns, yticklabels=corelation.columns,annot=True)
```

```python
sns.pairplot(Electric_Vehicle_Population_Data)
```

```python
sns.regplot(x="Model Year", y="Electric Range", data=Electric_Vehicle_Population_Data)
```

As the Model Year progresses' up,the Electric Range goes up:this indicates positive relationship between these two variable.

We can examine the correlation between "Electric Range" and "Model Year" and see its approximately -0.308
